```python
import os
import random

# Define the base project folder (where schools.html, images/, js/ are located)
# This script should be run from this 'your-project-folder'.
# If you run it from a different directory, adjust these paths accordingly.
base_project_dir = os.getcwd()

# Define the folders containing your media for ICTS SKA School
# These paths are relative to your 'your-project-folder'.
image_folder = os.path.join(base_project_dir, 'images', 'ICTS-SKA-School')
video_folder = os.path.join(base_project_dir, 'videos', 'ICTS-SKA-School')

# Common media extensions
image_extensions = ('.jpg', '.jpeg', '.png', '.webp', '.gif')
video_extensions = ('.mp4', '.webm', '.ogg', '.avi') # Add more video formats if needed

# List to store all media items as dictionaries {type: 'image'/'video', src: 'path', poster: 'path' (optional)}
all_media_items = []

# --- Helper function to process a specific media type folder ---
def process_media_type_folder(folder_path, media_type, extensions, base_src_prefix, new_file_prefix):
    """
    Processes a given folder for media files, renames them sequentially,
    and returns a list of structured media item dictionaries.
    """
    if not os.path.exists(folder_path):
        print(f"⚠️ Warning: Folder '{folder_path}' does not exist. Skipping {media_type} processing.")
        return []

    # Get all current media files (excluding directories)
    media_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f)) and f.lower().endswith(extensions)]

    if not media_files:
        print(f"⚠️ Warning: No {media_type} files found in '{folder_path}'.")
        return []

    # Step 1: Rename media to temporary names to avoid overwrite conflicts
    print(f"Renaming {len(media_files)} {media_type} files in '{folder_path}' to temporary names...")
    temp_media_paths = []
    for i, filename in enumerate(media_files):
        ext = os.path.splitext(filename)[1] # Get file extension (e.g., '.jpg', '.mp4')
        temp_name = f"temp_{new_file_prefix}_{i}{ext}" # Create a unique temporary name
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, temp_name)
        os.rename(old_path, new_path)
        temp_media_paths.append(new_path) # Store full path to temp file

    # Step 2: Rename temporary media files to final sequential names and create structured data
    structured_media = []
    print(f"Renaming temporary {media_type} files to final sequential names in '{folder_path}'...")
    for idx, temp_full_path in enumerate(temp_media_paths, 1):
        filename = os.path.basename(temp_full_path) # Get just the filename from temp_full_path
        ext = os.path.splitext(filename)[1]
        final_name = f"{new_file_prefix}{idx}{ext}" # e.g., 'image1.jpg' or 'video1.mp4'
        final_full_path = os.path.join(folder_path, final_name)
        os.rename(temp_full_path, final_full_path)

        # Construct the 'src' path that will be used in the HTML
        media_item = {'type': media_type, 'src': f'{base_src_prefix}/{final_name}'}

        # For videos, try to find a corresponding poster image in the image folder
        if media_type == 'video':
            video_base_name = os.path.splitext(final_name)[0] # e.g., 'video1'
            # Assumes poster image for 'video1.mp4' is 'video1.jpg' in the IMAGE folder
            potential_poster_file = f"{video_base_name}.jpg"
            potential_poster_full_path = os.path.join(os.path.dirname(image_folder), 'ICTS-SKA-School', potential_poster_file) # Look in the main image folder
            
            # This is the path for the HTML <img src="..."> relative to schools.html
            html_poster_path = f"images/ICTS-SKA-School/{potential_poster_file}"

            if os.path.exists(potential_poster_full_path):
                media_item['poster'] = html_poster_path
            else:
                print(f"Note: No poster found for video '{final_name}' at '{potential_poster_full_path}'.")


        structured_media.append(media_item)
    return structured_media

# --- Process Images ---
image_media_items = process_media_type_folder(
    folder_path=image_folder,
    media_type='image',
    extensions=image_extensions,
    base_src_prefix='images/ICTS-SKA-School', # HTML path prefix for images
    new_file_prefix='image'
)
all_media_items.extend(image_media_items)


# --- Process Videos ---
video_media_items = process_media_type_folder(
    folder_path=video_folder,
    media_type='video',
    extensions=video_extensions,
    base_src_prefix='videos/ICTS-SKA-School', # HTML path prefix for videos
    new_file_prefix='video'
)
all_media_items.extend(video_media_items)

# Shuffle the combined list of all media items (images and videos)
random.shuffle(all_media_items)

# --- Generate updated JavaScript list file ---
# The output JavaScript file path must precisely match the <script src="...">
# tag in your schools.html for ICTS SKA School.
output_js_file_path = os.path.join(base_project_dir, 'js', 'image_data', 'icts_ska_school_media.js')

# Ensure the directory structure for the JS file exists.
os.makedirs(os.path.dirname(output_js_file_path), exist_ok=True)

# Write the media paths to the JavaScript file.
print(f"Generating JavaScript file: {output_js_file_path}")
with open(output_js_file_path, "w") as js_file:
    # Write the global JavaScript variable declaration.
    # This variable name (window.ictsSkaSchoolMedia) MUST EXACTLY match
    # what's used in your schools.html's 'imagesMap'.
    js_file.write("window.ictsSkaSchoolMedia = [\n")
    for item in all_media_items:
        # Format the item as a JavaScript object literal
        if 'poster' in item:
            js_file.write(f"  {{ type: '{item['type']}', src: '{item['src']}', poster: '{item['poster']}' }},\n")
        else:
            js_file.write(f"  {{ type: '{item['type']}', src: '{item['src']}' }},\n")
    js_file.write("];\n") # Close the array and add a semicolon for proper JS syntax

print(f"✅ Renaming complete and {output_js_file_path} updated.")

```
